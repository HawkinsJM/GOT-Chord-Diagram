```python
output = []
for character in tv_char_list:
    for name in tv_char_list:
        if character is not name:
            stories = tvdf[(tvdf[character] == True) & (tvdf[name] == True)].count()['id']
            m_ratings = float(tvdf[(tvdf[character] == True) & (tvdf[name] == True) & (tvdf['rated'] == "M")].count()['id'])
            if stories > 0:
                m_ratio = m_ratings/stories
            else:
                m_ratio = 0
            output.append({"Character_A":character, "Character_B":name, "count":stories, "rating":m_ratio})
```

```python
output.sort(key=lambda x:(x['count'], x['rating']))
output.reverse()
```
